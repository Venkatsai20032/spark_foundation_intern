#Load and Inspect the Data

```python
import pandas as pd
df = pd.read_csv('/content/SampleSuperstore.csv')
print(df.head())
print(df.columns)
print(df.dtypes)
```

#Basic Statistical Analysis

```python
print(df.describe())
for col in df.select_dtypes(include=['object']).columns:
    print(f"{col}: {df[col].nunique()} unique values")
```

#Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
category_analysis = df.groupby('Category')[['Sales', 'Profit']].sum().reset_index()
sub_category_analysis = df.groupby('Sub-Category')[['Sales', 'Profit']].sum().reset_index()
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
sns.barplot(x='Category', y='Sales', data=category_analysis)
plt.title('Sales by Category')
plt.subplot(1, 2, 2)
sns.barplot(x='Category', y='Profit', data=category_analysis)
plt.title('Profit by Category')
plt.show()
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
sns.barplot(x='Sales', y='Sub-Category', data=sub_category_analysis.sort_values(by='Sales', ascending=False))
plt.title('Sales by Sub-Category')

plt.subplot(1, 2, 2)
sns.barplot(x='Profit', y='Sub-Category', data=sub_category_analysis.sort_values(by='Profit', ascending=False))
plt.title('Profit by Sub-Category')
plt.show()
```

#Regional Analysis

```python
region_analysis = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
sns.barplot(x='Region', y='Sales', data=region_analysis)
plt.title('Sales by Region')

plt.subplot(1, 2, 2)
sns.barplot(x='Region', y='Profit', data=region_analysis)
plt.title('Profit by Region')
plt.show()
```

#Segment Analysis

```python
segment_analysis = df.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
sns.barplot(x='Segment', y='Sales', data=segment_analysis)
plt.title('Sales by Segment')
plt.subplot(1, 2, 2)
sns.barplot(x='Segment', y='Profit', data=segment_analysis)
plt.title('Profit by Segment')
plt.show()
```

# Ship Mode Analysis

```python
ship_mode_analysis = df.groupby('Ship Mode')[['Sales', 'Profit']].sum().reset_index()
plt.figure(figsize=(14, 7))
plt.subplot(1, 2, 1)
sns.barplot(x='Ship Mode', y='Sales', data=ship_mode_analysis)
plt.title('Sales by Ship Mode')
plt.subplot(1, 2, 2)
sns.barplot(x='Ship Mode', y='Profit', data=ship_mode_analysis)
plt.title('Profit by Ship Mode')
plt.show()
```
